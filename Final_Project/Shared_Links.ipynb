# Working with Shared Links

Getting the File ID

* In your shared folder / file right click on the file and click on Share
* Copy link
* Here is an example of a copied link: https://drive.google.com/file/d/1-4d1PMlIyFgjWNnpPs_aD3gvVSGME07s/view?usp=drive_link
* Find the ID which in this case is 1-4d1PMlIyFgjWNnpPs_aD3gvVSGME07s
* Replace the id for the link provided below

```python
import pandas as pd

# https://drive.google.com/file/d/1-4d1PMlIyFgjWNnpPs_aD3gvVSGME07s/view?usp=drive_link
id = '1-4d1PMlIyFgjWNnpPs_aD3gvVSGME07s'
retail_data = pd.read_csv(f'https://docs.google.com/uc?id={id}')
retail_data.head()
```

```python
# create / connect to database
import sqlite3

connection = sqlite3.connect('ccpz.db') # creates / connects to db named after my team
cursor = connection.cursor()
```

```python
# df to sql
retail_data.to_sql('retail_data', connection, if_exists='fail')
```
